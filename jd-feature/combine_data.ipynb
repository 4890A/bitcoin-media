```python
import os
import glob
import pandas as pd
os.chdir("../csv")
```

```python
extension = 'csv'
all_filenames = [i for i in glob.glob('*.{}'.format(extension))]
```

```python
for f in all_filenames: 
    lines = open(f).readlines() 
    open(f, 'w').writelines(lines[1:]) 
```

```python
combined_csv = pd.concat([pd.read_csv(f) for f in all_filenames],  axis=1)

test = combined_csv.drop(["Month"], axis=1)
           
print(test)

#export to csv
test.to_csv( "combined_csv.csv", index=True, encoding='utf-8-sig')
```
