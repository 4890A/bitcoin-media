```python
c.executescript('''

    BEGIN TRANSACTION;
    /*create a new table with the same column names and types while
    defining a primary key for the desired column*/

    CREATE TABLE tweets_loc1 (location_id INTEGER PRIMARY KEY AUTOINCREMENT,
                              location TEXT,
                              latitude REAL,
                              longitude REAL);

    INSERT INTO tweets_loc1(location, latitude, longitude) SELECT location, latitude, longitude FROM tweets_loc;

    DROP TABLE tweets_loc;

    COMMIT TRANSACTION;''')

#close out the connection
c.close()
db.close()
```
